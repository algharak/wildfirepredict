```python
df_lightening = pd.read_csv(LIGHTENING_FILE)
print(df_lightening.to_string())
```

```python
df_lightening["count"] = 1
df_lightening["tested"] = 0
```

```python
print (df_lightening.columns)
row_ct = df_lightening.shape[0]
print (row_ct)
```

```python
lon_skew   =   2/360
lat_skew   =   1/180
```

```python
# This function calculates the upper and lower allowed limits of the latitude and longitude of a given coordinate.
# The percentage is set by lon_skew and lat_skew parameters set in prior cells
# Note: This approach will not work when the point of interest is at a borderline of coordinates (i.e.0 and 360 for the longitude and 0 and 180 for the latitude)

def get_edges (lt,lo):
    lt_min = lt - lat_skew
    lt_max = lt + lat_skew
    lo_min = lo - lon_skew
    lo_max = lo + lon_skew
    return lt_min,lt_max,lo_min,lo_max
```

```python
for row_num in range(row_ct-1):
    print(row_num)
    row_ser = df_lightening.iloc[row_num]
    if not row_ser ["tested"]:
        next_row = row_num + 1
        la_mi,la_ma,lo_mi,lo_ma = get_edges(row_ser["lat"],row_ser["lon"])
        for row_num_2 in range(next_row,row_ct):
            row_ser_2 = df_lightening.iloc[row_num_2]
            if not row_ser_2["tested"]:
                df_lightening.loc[df_lightening.index[row_num_2],"tested"] = 1
                if la_mi < row_ser_2["lat"] < la_ma and lo_mi < row_ser_2["lon"] <lo_ma:
                    print("found a match")
                    #row_ser ["count"] +=1
                    #df.loc[df.index[#], 'NAME']
                    df_lightening.loc[df_lightening.index[row_num],"count"] += 1

        #row_ser ["tested"] = 1
        #df_lightening.iloc[row_num] = row_ser
print(df_lightening.to_string())
```
